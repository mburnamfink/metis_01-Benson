```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

#import staring data
tb = pd.read_csv('Tidy_Benson.csv')
```

```python
#tidy up the data frame for pivoting
tb['D-T'] =  tb['WEEKDAY'].astype(str).map(lambda x : x.strip())+ '.' + tb['HOUR'].astype(str).map(lambda x: x.zfill(2))   
tr = tb[['STATION', 'D-T', 'PEOPLE']]
tr.pivot(index='STATION', columns='D-T', values='PEOPLE')
tr['PEOPLE'] = tr['PEOPLE'].map(lambda x : round(x)) #gives warning, okay to ignore
tr_heat_ready = tr.pivot(index='STATION', columns='D-T', values='PEOPLE')
```

```python
def make_heat_map(sns_data):
    """
    This method automates the heatmap generation of a dataframe. 
    Important internal parameters are size of output figure--currently a4_dims
    and cmap, a color pallet. Also not that the figure.tight_layout() needs to be
    the final call before saving the figure. The file is stored in the !pwd
    rather than renderd in jupyter which will sometimes crash when rendering
    large figures. 
    """
    a4_dims = (11.7, 8.27)
    fig, ax = plt.subplots(figsize=a4_dims)
    fig.set_size_inches(7, 4)
    sns_data = sns_data.head(20)
    sns_plot = sns.heatmap(data=sns_data,ax=ax, cmap='Greens', vmin=500, vmax=30000)
    figure = sns_plot.get_figure()
    figure.tight_layout()
    figure.savefig('svm_heat_test.png', dpi=400)
    
make_heat_map(my_data)  #stores as an external file due to rendering limits of jupyter
```
